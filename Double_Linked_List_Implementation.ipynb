```python
class Node:
    def __init__(self,student_id,name):
        self.student_id = student_id
        self.name = name
        self.prev = None
        self.next = None
```

```python
class LinkedList:
    def __init__(self):
        self.Head=None
        self.Tail=None
        #Add node at the end of the linked list
    def Add(self,student_id,name):
            nd=Node(student_id,name)
            #empty linked list
            if(self.Head==None):
                self.Head=nd
                self.Tail=nd
            else:
                nd.prev=self.Tail
                self.Tail.next=nd
                self.Tail=nd

        #insert a node of specific location
    def Insert(self,student_id, name,loc):
            nd=Node(student_id, name)  
            #empty linked list (loc has no effect)  
            if (self.Head==None):
                self.Head=nd
                self.Tail=nd
            #linked list has node
            else:
                if(loc==0):
                    nd.next=self.Head
                    self.Head.prev=nd
                    self.Head=nd
                else:
                    I=0
                    cur=self.Head
                    while(I<loc-1 and cur!=None):
                        cur=cur.next 
                        I=I+1  
                    if(cur==self.Tail or cur==None):
                            nd.prev=self.Tail
                            self.Tail.next=nd
                            self.Tail=nd
                        # node in middle
                    else:
                            cur.next.prev=nd
                            nd.next=cur.next
                            cur.next=nd
                            nd.prev=cur
       #search for a node by name
    def Search(self,name):
            nd=None
            #list is not empty
            if(self.Head!=None):
                nd=self.Head
                while(nd.name !=name and  nd!=None):
                    nd=nd.next
            return nd

        #delete a node at specific id
    def Delete(self,student_id):
            retval=False
            nd=None 
            #linkedlist is not empty
            if(self.Head != None):
                nd=self.Head
                #first node
                if(nd.student_id==student_id):
                # only 1 node
                    if(self.Head==self.Tail):
                        self.Head=self.Tail=None
                #more than one node
                    else:
                        self.Head=self.Head.next
                        self.Head.prev=None
                    return retval==True
                
                
                #not first node(middle or last)   
                else:
                    
                    while(nd.student_id != student_id and nd!=None):
                        
                        nd=nd.next
                        
                        #there is a node to delete
                    if(nd!=None):
                            #last node
                        if(nd==self.Tail):
                            self.Tail=self.Tail.prev
                            self.Tail.next=None
                            #middle node
                        else:  
                            nd.prev.next=nd.next
                            nd.next.prev=nd.prev
                        return retval==True
            return retval    
```

```python
class Menu:
    def __init__(self):
        self.linked_list = LinkedList()

    def display_menu(self):
        print("Linked List Menu:")
        print("1. Add student")
        print("2. Delete student")
        print("3. Search for a student")
        print("4. Insert student")
        print("5. Display student list")
        print("6. Exit")

    def implement_menu(self):
        while True:
            self.display_menu()
            choice = input("Enter your choice: ")

            if choice == "1":
                student_id = input("Enter student ID: ")
                name = input("Enter student name: ")
                self.linked_list.Add(student_id, name)
                print("Student added to the end of the linked list.")

            elif choice == "2":
                student_id = input("Enter the student ID to delete: ")
                if self.linked_list.Delete(student_id):
                    print("Student deleted successfully.")
                else:
                    print("Student not found or deletion failed.")

            elif choice == "3":
                name = input("Enter the student Name to search for: ")
                node = self.linked_list.Search(name)
                if node:
                    print(f"Student found: ID - {node.student_id}, Name - {node.name}")
                else:
                    print("Student not found.")

            elif choice == "4":
                student_id = input("Enter student ID: ")
                name = input("Enter student name: ")
                loc = int(input("Enter the index where you want to insert the student: "))
                self.linked_list.Insert(student_id, name, loc)
                print("Student inserted successfully.")

            elif choice == "5":
                print("Student List:")
                current = self.linked_list.Head
                while current:
                    print(f"ID: {current.student_id}, Name: {current.name}")
                    current = current.next

            elif choice == "6":
                print("Exiting from this program.")
                break

            else:
                print("Invalid choice. Please enter a valid option.")

```
