```python
class Queue:
    def __init__(self):
        self.queue = [None] * 5
        self.head = self.tail = -1

    def enqueue(self, item):
        if self.tail  == (len(self.queue)-1) and self.head==0:
            print("Queue is full")
            return False

        if (self.head == -1):
            self.head = self.tail = 0
        else:
            self.tail = self.tail + 1 

        self.queue[self.tail] = item
        return True

    def dequeue(self):
        if (self.head == -1 and self.tail==-1):
            print("Queue is empty")
            return None

        d = self.queue[self.head]
        self.head = self.head + 1
        if(self.head==len(self.queue) ):
            self.head = self.tail = -1

        return d   
```

```python
class Menu:
    def __init__(self):
        self.queue = Queue()

    def display_menu(self):
        print("\nQueue Menu:")
        print("1. Add number")
        print("2. Display All")
        print("3. Delete number")
        print("4. Free Queue")
        print("5. Exit")

    def implement_menu(self):
        while True:
            self.display_menu()
            choice = input("Enter your choice: ")

            if choice == "1":
                number = input("Enter the number to add: ")
                self.queue.enqueue(number)
                
            elif choice == "2":
                print("Queue contents:")
                current_head = self.queue.head
                while current_head != len(self.queue.queue):
                    if (self.queue.queue[current_head] != None):
                        print(self.queue.queue[current_head])
                   
                    current_head = current_head + 1
                    
            elif choice == "3":
                dequeued_number = self.queue.dequeue()
                if (dequeued_number != None):
                    print("Deleted number:", dequeued_number)
                    
            elif choice == "4":
                current_head = self.queue.head
                
                while  current_head != len(self.queue.queue):
                    self.queue.dequeue()
                    current_head = current_head + 1
                    
                print("Queue has been freed.")
                
            elif choice == "5":
                print("Exiting the program.")
                break
            else:
                print("Invalid choice. Please enter a valid option.")
```

```python
if __name__ == "__main__":
    menu = Menu()
    menu.implement_menu()
```
