```python
class Queue:
    def __init__(self):
        
        self.arr = [0] * 5
        self.front = -1
        self.rear = -1

    def is_empty(self):
        return self.front == -1

    def is_full(self):
        return (self.front == 0 and self.rear == 4) or (self.rear == (self.front - 1))

    def enqueue(self, d):
        if self.is_full():
            print("Queue is full")
            return False

        if self.front == -1:
            self.front = self.rear = 0
        else:
            self.rear = (self.rear + 1) % 5

        self.arr[self.rear] = d
        return True

    def dequeue(self):
        if self.is_empty():
            print("Queue is empty")
            return  None

        d = self.arr[self.front]
        if self.front == self.rear:
            self.front = self.rear = -1
        else:
            self.front = (self.front + 1) % 5

        return d

```

```python
class StackMenu:
    def __init__(self):
        self.queue = Queue()

    def display_menu(self):
        print("\nQueue Menu:")
        print("1. Add number")
        print("2. Display All")
        print("3. Delete number")
        print("4. Free Queue")
        print("5. Exit")

    def execute_menu(self):
        while True:
            self.display_menu()
            choice = input("Enter your choice: ")

            if choice == "1":
                d = input("Enter integer number: ")
                self.queue.enqueue(d)
                print("Number added to the queue.")

            elif choice == "2":
                current = self.queue.front
                print("Queue contents:")
                while current != -1:
                    print(f"Number: {self.queue.arr[current]}")
                    current = (current + 1) % 5
                    if current == (self.queue.rear + 1) % 5:
                        break


            elif choice == "3":
                d_deleted = self.queue.dequeue()
                if d_deleted:
                    print(f"Number dequeued: {d_deleted}")
                else:
                    print("Queue is empty.")

            elif choice == "4":
                while not self.queue.is_empty():
                    n_dequeued = self.queue.dequeue()
                    print(f"Number dequeued: {n_dequeued}")
                print("Queue freed. All students removed.")

            elif choice == "5":
                print("Exiting the program.")
                break

            else:
                print("Invalid choice. Please enter a valid option.")
```

```python
menu = StackMenu()
menu.execute_menu()
```
