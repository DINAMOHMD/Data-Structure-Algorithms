```python
class Node:
    def __init__(self,student_id,name):
        self.student_id = student_id
        self.name = name
        self.prev = None
        
```

```python
class StackList:
    def __init__(self):
        self.Tail = None

    def push(self, student_id, name):
        nd = Node(student_id, name)
        if self.Tail is None:
            self.Tail = nd
        else:
            nd.prev = self.Tail
            self.Tail = nd

    def pop(self):
        nd = self.Tail
        if self.Tail is not None:
            self.Tail = self.Tail.prev
        return nd
```

```python
class StackMenu:
    def __init__(self):
        self.stack = StackList()

    def display_menu(self):
        print("\nStack Menu:")
        print("1. Add student")
        print("2. Display All")
        print("3. Delete student")
        print("4. Free Stack")
        print("5. Exit")

    def execute_menu(self):
        while True:
            self.display_menu()
            choice = input("Enter your choice: ")

            if choice == "1":
                student_id = input("Enter student ID: ")
                name = input("Enter student name: ")
                self.stack.push(student_id, name)
                print("Student added to the stack.")

            elif choice == "2":
                current = self.stack.Tail
                print("Stack contents:")
                while current:
                    print(f"Student ID: {current.student_id}, Name: {current.name}")
                    current = current.prev


            elif choice == "3":
                deleted_student = self.stack.pop()
                if deleted_student:
                    print(f"Student deleted: ID - {deleted_student.student_id}, Name - {deleted_student.name}")
                else:
                    print("Student not found in the stack.")

            
            elif choice == "4":
                while (self.stack.Tail != None):
                    popped_student = self.stack.pop()
                    print(f"Student popped: ID - {popped_student.student_id}, Name - {popped_student.name}")
                print("Stack freed. All students removed.")



            elif choice == "5":
                print("Exiting the program.")
                break

            else:
                print("Invalid choice. Please enter a valid option.")
```

```python
if __name__ == "__main__":
    stack_menu = StackMenu()
    stack_menu.execute_menu()
```
